```python
# Exemplo de conversão de uma imagem em RGB para XYZ.

from PIL import Image
import numpy as np

# Tamanho da imagem
width, height = 400, 300

# Criar uma imagem em RGB com fundo aleatório
rgb_image = Image.new("RGB", (width, height))
rgb_pixels = rgb_image.load()

# Preencher a imagem RGB com cores aleatórias
for x in range(width):
    for y in range(height):
        r, g, b = np.random.randint(0, 256, 3)
        rgb_pixels[x, y] = (int(r), int(g), int(b))

# Salvar a imagem RGB
rgb_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/xyz/rgb_image_with_random_background.jpg")

# Converter a imagem RGB para XYZ (simplificado)
def rgb_to_xyz(rgb):
    r, g, b = rgb
    x = 0.4124564 * r + 0.3575761 * g + 0.1804375 * b
    y = 0.2126729 * r + 0.7151522 * g + 0.0721750 * b
    z = 0.0193339 * r + 0.1191920 * g + 0.9503041 * b
    return x, y, z

xyz_image = Image.new("RGB", (width, height))
xyz_pixels = xyz_image.load()

# Converter pixels RGB para XYZ e aplicar à nova imagem
for x in range(width):
    for y in range(height):
        r, g, b = rgb_pixels[x, y]
        x_val, y_val, z_val = rgb_to_xyz((r, g, b))
        xyz_pixels[x, y] = (int(x_val), int(y_val), int(z_val))

# Salvar a imagem XYZ
xyz_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/xyz/xyz_image.jpg")
```

```python
# Exemplo de conversão de uma imagem em XYZ para RGB.

import numpy as np
from PIL import Image

# Tamanho da imagem
width, height = 400, 300

# Gerar uma imagem aleatória no espaço de cores XYZ
xyz_image = np.random.rand(height, width, 3) * 100  # Valores XYZ aleatórios

# Definir a matriz de conversão de XYZ para RGB
xyz_to_rgb_matrix = np.array([
    [3.2406, -1.5372, -0.4986],
    [-0.9689, 1.8758, 0.0415],
    [0.0557, -0.2040, 1.0570]
])

# Converter de XYZ para RGB
xyz_image_reshaped = xyz_image.reshape(-1, 3)
rgb_image_reshaped = np.dot(xyz_image_reshaped, xyz_to_rgb_matrix.T)
rgb_image_reshaped = np.clip(rgb_image_reshaped, 0, 1)  # Garantir que os valores RGB estejam no intervalo [0, 1]
rgb_image = (rgb_image_reshaped * 255).astype(np.uint8).reshape(height, width, 3)

# Criar uma imagem PIL a partir dos valores RGB
rgb_image_pil = Image.fromarray(rgb_image)

# Salvar a imagem RGB
rgb_image_pil.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/xyz/random_xyz_to_rgb.jpg")

# Salvar a imagem XYZ (opcional)
xyz_image_pil = Image.fromarray((xyz_image / 100 * 255).astype(np.uint8))
xyz_image_pil.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/xyz/random_xyz_image.jpg")
```

```python
# Exemplo de conversão de imagem RGB em YIQ.

from PIL import Image
import numpy as np

# Tamanho da imagem
width, height = 400, 300

# Criar uma imagem em RGB com fundo aleatório
rgb_image = Image.new("RGB", (width, height))
rgb_pixels = rgb_image.load()

# Preencher a imagem RGB com cores aleatórias
for x in range(width):
    for y in range(height):
        r, g, b = np.random.randint(0, 256, 3)
        rgb_pixels[x, y] = (int(r), int(g), int(b))

# Salvar a imagem RGB
rgb_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/yiq/rgb1_image_with_random_background.jpg")

# Converter a imagem RGB para YIQ
def rgb_to_yiq(rgb):
    r, g, b = rgb
    y = 0.299 * r + 0.587 * g + 0.114 * b
    i = 0.596 * r - 0.274 * g - 0.322 * b
    q = 0.211 * r - 0.523 * g + 0.312 * b
    return y, i, q

yiq_image = Image.new("RGB", (width, height))
yiq_pixels = yiq_image.load()

# Converter pixels RGB para YIQ e aplicar à nova imagem
for x in range(width):
    for y in range(height):
        r, g, b = rgb_pixels[x, y]
        y_val, i_val, q_val = rgb_to_yiq((r, g, b))
        yiq_pixels[x, y] = (int(y_val), int(i_val), int(q_val))

# Salvar a imagem YIQ
yiq_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/yiq/yiq_image.jpg")
```

```python
# Exemplo de conversão de imagem YIQ em RGB.

import numpy as np
from PIL import Image

# Tamanho da imagem
width, height = 400, 300

# Gerar uma imagem aleatória no espaço de cores YIQ
yiq_image = np.random.rand(height, width, 3) * 255  # Valores YIQ aleatórios

# Matriz de conversão de YIQ para RGB
yiq_to_rgb_matrix = np.array([
    [1.0, 0.956, 0.621],
    [1.0, -0.272, -0.647],
    [1.0, -1.108, 1.705]
])

# Converter de YIQ para RGB
yiq_image_reshaped = yiq_image.reshape(-1, 3)
rgb_image_reshaped = np.dot(yiq_image_reshaped, yiq_to_rgb_matrix.T)
rgb_image_reshaped = np.clip(rgb_image_reshaped, 0, 255).astype(np.uint8)
rgb_image = rgb_image_reshaped.reshape(height, width, 3)

# Criar uma imagem PIL a partir dos valores RGB
rgb_image_pil = Image.fromarray(rgb_image)

# Salvar a imagem RGB
rgb_image_pil.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/yiq/random_yiq_to_rgb.jpg")

# Salvar a imagem YIQ (opcional)
yiq_image_pil = Image.fromarray(yiq_image.astype(np.uint8))
yiq_image_pil.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/yiq/random_yiq_image.jpg")
```

```python
# Exemplo de conversão de uma imagem YCrCb para RBG

import numpy as np
from PIL import Image

# Tamanho da imagem
width, height = 400, 300

# Gerar uma imagem aleatória no espaço de cores YCrCb
ycrcb_image = np.random.rand(height, width, 3) * 255  # Valores Y, Cr, Cb aleatórios

# Matrizes de conversão de YCrCb para RGB
ycrcb_to_rgb_matrix = np.array([
    [1.0, 1.0, 0.0],
    [1.0, -0.34414, -0.71414],
    [1.0, 1.772, 0.0]
])

# Aplicar as matrizes de conversão
rgb_image = np.zeros_like(ycrcb_image)
for i in range(height):
    for j in range(width):
        y, cr, cb = ycrcb_image[i, j]
        rgb_image[i, j] = np.dot(ycrcb_to_rgb_matrix, [y, cr - 128, cb - 128])

# Garantir que os valores RGB estejam no intervalo [0, 255]
rgb_image = np.clip(rgb_image, 0, 255).astype(np.uint8)

# Criar uma imagem PIL a partir dos valores RGB
rgb_image_pil = Image.fromarray(rgb_image)

# Salvar a imagem RGB
rgb_image_pil.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/ycrcb/random_ycrcb_to_rgb.jpg")

# Salvar a imagem YCrCb (opcional)
ycrcb_image_pil = Image.fromarray(ycrcb_image.astype(np.uint8))
ycrcb_image_pil.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/ycrcb/random_ycrcb_image.jpg")
```

```python
#===============================================================================

# Operações lógicas em imagens - dia 01/03/2024

#===============================================================================

from PIL import Image, ImageDraw

# Tamanho da imagem
width, height = 400, 300

# Criar a primeira imagem com fundo branco e objeto preto
image1 = Image.new("1", (width, height), 1)  # Modo "1" é binário (preto e branco)
draw1 = ImageDraw.Draw(image1)
draw1.rectangle([50, 50, 200, 200], fill=0)  # Desenha um retângulo preto
draw1.line((0, 0, width, height), fill=0)


# Criar a segunda imagem com fundo branco e objeto preto em uma posição diferente
image2 = Image.new("1", (width, height), 1)
draw2 = ImageDraw.Draw(image2)
draw2.rectangle([250, 100, 350, 250], fill=0)

# Salvar as duas imagens iniciais
image1.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/logica/image1.png")
image2.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/logica/image2.png")

# Realizar a operação "and" pixel a pixel
image_and = Image.new("1", (width, height), 1)
for x in range(width):
    for y in range(height):
        pixel1 = image1.getpixel((x, y))
        pixel2 = image2.getpixel((x, y))
        result_pixel = pixel1 | pixel2
        image_and.putpixel((x, y), result_pixel)

# Salvar a imagem resultante da operação "and"
image_and.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/logica/image_and.png")

# Realizar a operação "or" pixel a pixel
image_or = Image.new("1", (width, height), 1)
for x in range(width):
    for y in range(height):
        pixel1 = image1.getpixel((x, y))
        pixel2 = image2.getpixel((x, y))
        result_pixel = pixel1 & pixel2
        image_or.putpixel((x, y), result_pixel)

# Salvar a imagem resultante da operação "or"
image_or.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/logica/image_or.png")

# Realizar a operação "nand" pixel a pixel
image_nand = Image.new("1", (width, height), 1)
for x in range(width):
    for y in range(height):
        pixel1 = image1.getpixel((x, y))
        pixel2 = image2.getpixel((x, y))
        result_pixel = not(pixel1 | pixel2)
        image_nand.putpixel((x, y), result_pixel)

# Salvar a imagem resultante da operação "or"
image_nand.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/logica/image_nand.png")
```

```python
#===============================================================================

# Histograma da imagem spider - dia 08/03/24

#===============================================================================

from PIL import Image
import matplotlib.pyplot as plt

# Carregue a imagem
image = Image.open('/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/histograma/05 - folhas.jpg')
width, height = image.size

# Converta a imagem para o modo RGB (para garantir que tenha todas as três componentes)
image_rgb = image.convert('RGB')

# Converta a imagem para escala de cinza <<<<<<<<<<
image_gray = image.convert('L')

# Extraia as bandas de cores
r, g, b = image_rgb.split()

# Calcule os histogramas para cada componente
hist_r = r.histogram()
hist_g = g.histogram()
hist_b = b.histogram()

# Calcule o histograma da escala de cinza <<<<<<<<<
hist_gray = image_gray.histogram()

# Plotar os histogramas
plt.figure(figsize=(18, 6))

plt.subplot(141)
plt.plot(hist_r, color='red')
plt.title('Histograma da Componente R')
plt.xlim([0, 256])

plt.subplot(142)
plt.plot(hist_g, color='green')
plt.title('Histograma da Componente G')
plt.xlim([0, 256])

plt.subplot(143)
plt.plot(hist_b, color='blue')
plt.title('Histograma da Componente B')
plt.xlim([0, 256])

plt.subplot(144)
plt.plot(hist_gray, color='gray')
plt.title('Histograma da Escala de Cinza')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
#===============================================================================

# Binarização da Imagem - dia 15/03/24

#===============================================================================

from PIL import Image

# Carregue a imagem em escala de cinza
image_gray = Image.open('/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/histograma/folhas_gray.jpg')

# Defina um limiar (valor entre 0 e 255)
limiar = 170

# Obtenha as dimensões da imagem
largura, altura = image_gray.size

# Crie uma nova imagem para armazenar a imagem binarizada
image_binarized = Image.new('L', (largura, altura))

# Binarize a imagem pixel a pixel
for y in range(altura):
    for x in range(largura):
        pixel = image_gray.getpixel((x, y))
        if pixel < limiar:
            novo_pixel = 0  # Preto
        else:
            novo_pixel = 255  # Branco
        image_binarized.putpixel((x, y), novo_pixel)

# Salve a imagem binarizada
image_binarized.save('/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/histograma/folhas_binarizada_170.jpg')
```

```python
#===============================================================================

# Equalização de Histograma - dia 15/03/24

#===============================================================================

# biblioteca PIL
from PIL import Image

# adquirindo imagem no mesmo diretorio
image_file = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/histograma/wolverine_sc.jpg"
img_orig = Image.open(image_file)

# pegando tamanho da imagem
width, height = img_orig.size

# inicializando variaveis:
totalpixels = width * height
constsomatorio = 255.0/totalpixels
vetorR = []
vetorS = []
for i in range(0,256):
	vetorR.append([])
	vetorS.append([])

for i in range(0, 256):
  vetorR[i]=0

#criando as novas imagens com o tamanho da anterior
img_novacinza = Image.new("RGB", img_orig.size)
img_novaequalizada = Image.new("RGB", img_orig.size)

# laco que corre pixel a pixel modificando para escala de cinza, vai ser utilizado para a imagem em escala de cinza e a confeccao do histograma.
# pega o valor RGB, 012, soma e divide por 3, jogando para a variavel valor.
# no vetorR[0..255], a posicao do vetor representa a cor e o valor eh a quantidade dessa cor na imagem, inicialmente o vetor foi inicializado com 0.
for x in range(img_orig.size[0]):
        for y in range(img_orig.size[1]):
                valor = int((img_orig.getpixel((x,y))[0] + img_orig.getpixel((x,y))[1] + img_orig.getpixel((x,y))[2])/3)
                img_novacinza.putpixel((x,y), (valor,valor,valor)) #nova imagem em escala de cinza
                vetorR[valor]=vetorR[valor]+1 #fazendo histograma no vetorR[0..255] a quantidade eh incrementada toda vez que a cor eh identificada.

#lacos que geram o vetorS[0..255] por resultado do somatorio das posicoes do vetorR multiplicado pela constante constsomatorio calculada anteriormente.
#no vetorS[0..255], a posicao do vetor representa a cor antiga e o valor eh a cor nova.
vetorS[0]=vetorR[0]*constsomatorio
for i in range(1,256):
  vetorS[i]=vetorS[i-1]+vetorR[i]*constsomatorio

#neste ponto a variavel vetorS eh do tipo float por heranca da variavel constsomatorio e deve ser convertida para int, ja recebendo a aproximacao que
#o proprio python faz na conversao.
#sem este passo, nao seria possivel entrar valores na funcao "putpixel"
for i in range(0,256):
        vetorS[i] = int (vetorS[i])

#lacos que vao equalizar a imagem, eh feita a leitura da imagem pixel a pixel, com a cor do pixel, vamos ao vetorS[0..255] e adicionamos a nova cor.
for x in range(img_novacinza.size[0]):
        for y in range(img_novacinza.size[1]):
                valor = (img_novacinza.getpixel((x,y))[0])
                img_novaequalizada.putpixel((x,y), (vetorS[valor],vetorS[valor],vetorS[valor]))

#salva em novos arquivos
img_novacinza.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/histograma/wolverine_sc_ec.jpg")
img_novaequalizada.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/histograma/wolverine_equalizada.jpg")
```

```python
#===============================================================================

# Aplicação do Filtro Média - dia 22/03/24

#===============================================================================

# importando bibliotecas
from PIL import Image
from numpy import *

# recebendo imagem
#figura = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_media/spider.jpg"
figura = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_media/folhas.jpg"
image_file = figura
img = Image.open(image_file)

# inicializando variaveis
width, height = img.size
totalpixels = height * width # total de pixels da imagem, serve de indice para vetores e matrizes

# criando matrizes com o numpy
A = arange(totalpixels) #total de elementos
A.shape = (width,height) #numpy.ndarray no formato da figura
corR = mat(A.copy()) #numpy.core.defmatrix.matrix no formato da figura
corG = mat(A.copy())
corB = mat(A.copy())
corEC = mat(A.copy())
HR = mat(A.copy())
HG = mat(A.copy())
HB = mat(A.copy())
HEC = mat(A.copy())

# inicializa matrizes que serao utilizadas para calculos de H em RGB
for x in range(img.size[0]):
    for y in range(img.size[1]):
        corR[x,y] = img.getpixel((x,y))[0]
        corG[x,y] = img.getpixel((x,y))[1]
        corB[x,y] = img.getpixel((x,y))[2]
        corEC[x,y] = (img.getpixel((x,y))[0]+img.getpixel((x,y))[1]+img.getpixel((x,y))[2])/3


# criando novas imagens RGB para o restante do processamento
imgmedia = Image.new("RGB", img.size)
imgmediaEC = Image.new("RGB", img.size)

# calculo de H para RGB
for x in range(1,width-1):
    for y in range(1,height-1):
        HR[x,y]=(corR[x-1,y-1]+corR[x,y-1]+corR[x+1,y-1]+corR[x-1,y]+corR[x,y]+corR[x+1,y]+corR[x-1,y+1]+corR[x,y+1]+corR[x+1,y+1])/9
        HG[x,y]=(corG[x-1,y-1]+corG[x,y-1]+corG[x+1,y-1]+corG[x-1,y]+corG[x,y]+corG[x+1,y]+corG[x-1,y+1]+corG[x,y+1]+corG[x+1,y+1])/9
        HB[x,y]=(corB[x-1,y-1]+corB[x,y-1]+corB[x+1,y-1]+corB[x-1,y]+corB[x,y]+corB[x+1,y]+corB[x-1,y+1]+corB[x,y+1]+corB[x+1,y+1])/9
        HEC[x,y]=(corEC[x-1,y-1]+corEC[x,y-1]+corEC[x+1,y-1]+corEC[x-1,y]+corEC[x,y]+corEC[x+1,y]+corEC[x-1,y+1]+corEC[x,y+1]+corEC[x+1,y+1])/9


# monta imagem
for x in range(img.size[0]):
    for y in range(img.size[1]):
        if x == 0 or y == 0 or x == img.size[0] or y == img.size[1]:
            imgmedia.putpixel((x,y), (corR[x,y],corG[x,y],corB[x,y]))
            imgmediaEC.putpixel((x,y), (corEC[x,y],corEC[x,y],corEC[x,y]))
        else:
            imgmedia.putpixel((x,y), (HR[x,y],HG[x,y],HB[x,y]))
            imgmediaEC.putpixel((x,y), (HEC[x,y],HEC[x,y],HEC[x,y]))


# salvando imagem
#imgmedia.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_media/spider_media.jpg")
#imgmediaEC.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_media/spider_media_EC.jpg")
imgmedia.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_media/folhas_media.jpg")
imgmediaEC.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_media/folhas_media_EC.jpg")
```

```python
#===============================================================================

# Aplicação do Filtro Roberts - dia 22/03/24

#===============================================================================

from PIL import Image, ImageFilter

# Função para aplicar o filtro de Roberts usando convolução
def filtro_roberts(imagem):
    # Converte a imagem para escala de cinza (opcional)
    imagem_cinza = imagem.convert("L")

    # Aplica o filtro de Roberts
    imagem_filtrada = imagem_cinza.filter(ImageFilter.FIND_EDGES)

    return imagem_filtrada

# Carrega a imagem "spider.jpg"
#imagem = Image.open("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_roberts/spider.jpg")
imagem = Image.open("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_roberts/folhas.jpg")

# Aplica o filtro de Roberts usando a função
imagem_filtrada = filtro_roberts(imagem)

# Exibe a imagem original e a imagem filtrada (opcional)
imagem.show()
imagem_filtrada.show()

# Salva a imagem filtrada (opcional)
#imagem_filtrada.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_roberts/spider_roberts.jpg")
imagem_filtrada.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_roberts/folhas_roberts.jpg")
```

```python
#===============================================================================

# Aplicação do Filtro Sobel - dia 22/03/24

#===============================================================================

# importando bibliotecas
from PIL import Image
from numpy import *

# recebendo limiar
limiar = int(input("Limiar (0-255):"))

# recebendo imagem
#figura = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_sobel/spider.jpg"
figura = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_sobel/folhas.jpg"
image_file = figura
img = Image.open(image_file)

# inicializando variaveis
width, height = img.size
totalpixels = height * width # total de pixels da imagem, serve de indice para vetores e matrizes

# criando matrizes com o numpy
A = arange(totalpixels) #total de elementos
A.shape = (width,height) #numpy.ndarray no formato da figura
corEC = mat(A.copy()) #numpy.core.defmatrix.matrix no formato da figura
HEC = mat(A.copy())
LIM = mat(A.copy())

# inicializa matrizes que serao utilizadas para calculos de H em RGB
for x in range(img.size[0]):
    for y in range(img.size[1]):
        corEC[x,y] = (img.getpixel((x,y))[0] + img.getpixel((x,y))[1] + img.getpixel((x,y))[2])/3

# criando novas imagens RGB para o restante do processamento
imgsobel = Image.new("RGB", img.size)

# calculo de H para RGB
for x in range(1,width-1):
    for y in range(1,height-1):
        # calcula R1 em Escala de Cinza
        R1R=(-1*corEC[x-1,y-1])+(-2*corEC[x,y-1])+(-1*corEC[x+1,y-1])+(0*corEC[x-1,y])+(0*corEC[x,y])+(0*corEC[x+1,y])+(1*corEC[x-1,y+1])+(2*corEC[x,y+1])+(1*corEC[x+1,y+1])

        # calcula R2 em Escala de Cinza
        R2R=(-1*corEC[x-1,y-1])+(0*corEC[x,y-1])+(1*corEC[x+1,y-1])+(-2*corEC[x-1,y])+(0*corEC[x,y])+(2*corEC[x+1,y])+(-1*corEC[x-1,y+1])+(0*corEC[x,y+1])+(1*corEC[x+1,y+1])

        # calcula H de R1 e R2 em Escala de Cinza
        HEC[x,y]=sqrt(R1R**2+R2R**2)

# aplica limiar
for x in range(1,width-1):
    for y in range(1,height-1):
        if HEC[x,y] > limiar:
            LIM[x,y] = 255
        else:
            LIM[x,y] = 0

# monta imagem
for x in range(img.size[0]):
    for y in range(img.size[1]):
        imgsobel.putpixel((x,y), (LIM[x,y],LIM[x,y],LIM[x,y]))

# salvando imagem
limiar = str(limiar)
#nome = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_sobel/spider_sobel_l_"+limiar+".jpg"
nome = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/f_sobel/folhas_sobel_l_"+limiar+".jpg"
imgsobel.save(nome)
```

```python
#===============================================================================

# Transformação 2D - Rotação - dia 05/04/2024

#===============================================================================

# importando bibliotecas
from PIL import Image
from numpy import *
from math import *

# recebendo imagem
figura = "/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/t_rotacao/quadrado.bmp"
image_file = figura
img = Image.open(image_file)

# calculos da imagem
width, height = img.size
totalpixels = width*height

# inicializando variaveis
maxx = 0 # coordenada x onde esta o ultimo pixel da figura em preto
minx = 3000 # coordenada x onde esta o primeiro pixel da figura em preto
maxy = 0 # coordenada y onde esta o ultimo pixel da figura em preto
miny = 3000 # coordenada y onde esta o primeiro pixel da figura em preto

# criando matrizes com o numpy
A = arange(totalpixels) #total de elementos
A.shape = (width,height) #numpy.ndarray no formato da figura
PixFig = mat(A.copy()) #numpy.core.defmatrix.matrix no formato da figura
AuxImagem = mat(A.copy())
NovaImagem = mat(A.copy())

# principal
# inicializa matrizes
for x in range(img.size[0]):
    for y in range(img.size[1]):
        if img.getpixel((x,y)) == 255:
                PixFig[x,y] = 1 # matriz que contem a imagem marcada 1 --> fundo, 0 --> figura
        else:
                PixFig[x,y] = 0 # matriz que contem a imagem marcada 1 --> fundo, 0 --> figura
        AuxImagem[x,y] = img.getpixel((x,y)) # matriz copia da imagem
        NovaImagem[x,y] = 255 # matriz que representa o fundo da nova imagem

# marcando maximos e minimos
for y in range(img.size[1]):
   for x in range(img.size[0]):
        if (PixFig[x,y]-PixFig[x-1,y]) == -1 and x < minx:
                minx = x
        if (PixFig[x,y]-PixFig[x-1,y]) == 1 and maxx < x:
                maxx = x
        if (PixFig[x,y]-PixFig[x,y-1]) == -1 and y < miny:
                miny = y
        if (PixFig[x,y]-PixFig[x,y-1]) == 1 and maxy < y:
                maxy = y

# diagonal da figura demarcada na imagem
diagonal = int(sqrt((maxx-minx)**2 + (maxy-miny)**2)) + 2

# baricentro da figura em relacao a x=0 e y=0 da imagem
bx = ((maxx-minx)/2)+minx
by = ((maxy-miny)/2)+miny

# intervalo de variacao de x e y para rotacionar a figura dentro da imagem
intminx = int(-(diagonal/2)+bx)
intmaxx = int((diagonal/2)+bx)
intminy = int(-(diagonal/2)+by)
intmaxy = int((diagonal/2)+by)

# receber angulo de rotacao
angulo = int(input("Angulo: "))
angulorad = (angulo/180.0)*pi # angulo convertido em radianos para operacoes dentro do python

# criando novas imagens RGB para o restante do processamento
imgrot = Image.new("RGB", img.size)

# formacao da imagem de rotacao
for x in range(intminx, intmaxx):
    for y in range(intminy, intmaxy):
        # novos pontos para x e y de acordo com o angulo informado
        novo_x = int(x * cos(angulorad) - y * sin(angulorad))
        novo_y = int(x * sin(angulorad) + y * cos(angulorad))
        # matriz que contem nova imagem
        NovaImagem[novo_x,novo_y]=AuxImagem[x,y]

# plotando imagem nova em RGB
for x in range(img.size[0]):
    for y in range(img.size[1]):
        imgrot.putpixel((x,y), (NovaImagem[x,y],NovaImagem[x,y],NovaImagem[x,y]))

# salvando imagens
nome="/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/t_rotacao/imgrot_"+str(angulo)+".jpg"
imgrot.save(nome)
```

```python
# transformada e filtragem de imagem aleatória, salvando fft e filtros.

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Função para aplicar um filtro passa alta
def high_pass_filter(frequency_image, cutoff_frequency):
    rows, cols = frequency_image.shape
    center_row, center_col = rows // 2 , cols // 2
    mask = np.ones((rows, cols))
    mask[center_row - cutoff_frequency:center_row + cutoff_frequency,
         center_col - cutoff_frequency:center_col + cutoff_frequency] = 0
    return frequency_image * mask

# Função para aplicar um filtro passa baixa
def low_pass_filter(frequency_image, cutoff_frequency):
    rows, cols = frequency_image.shape
    center_row, center_col = rows // 2 , cols // 2
    mask = np.zeros((rows, cols))
    mask[center_row - cutoff_frequency:center_row + cutoff_frequency,
         center_col - cutoff_frequency:center_col + cutoff_frequency] = 1
    return frequency_image * mask

# Gerar uma imagem aleatória de 200x200 pixels
random_image = Image.fromarray(np.random.randint(0, 256, (200, 200), dtype=np.uint8))

# Converter a imagem para um array numpy
image_array = np.array(random_image, dtype=np.float32)

# Aplicar a Transformada de Fourier
frequency_image = np.fft.fftshift(np.fft.fft2(image_array))

# Salvar a Transformada de Fourier como imagem
magnitude_spectrum = np.log(np.abs(frequency_image) + 1)  # Usar log para melhor visualização
magnitude_spectrum /= magnitude_spectrum.max()  # Normalizar para 0-1
magnitude_spectrum *= 255
Image.fromarray(magnitude_spectrum.astype(np.uint8)).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/fourier_transform.png")

# Aplicar filtro passa alta e passa baixa
cutoff_frequency = 30
high_pass_result = high_pass_filter(frequency_image, cutoff_frequency)
low_pass_result = low_pass_filter(frequency_image, cutoff_frequency)

# Salvar as transformadas filtradas como imagens
high_pass_magnitude = np.log(np.abs(high_pass_result) + 1)
high_pass_magnitude /= high_pass_magnitude.max()
high_pass_magnitude *= 255
Image.fromarray(high_pass_magnitude.astype(np.uint8)).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/high_pass_transform.png")

low_pass_magnitude = np.log(np.abs(low_pass_result) + 1)
low_pass_magnitude /= low_pass_magnitude.max()
low_pass_magnitude *= 255
Image.fromarray(low_pass_magnitude.astype(np.uint8)).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/low_pass_transform.png")

# Aplicar a Transformada Inversa de Fourier para voltar ao domínio do espaço
high_pass_image = np.abs(np.fft.ifft2(np.fft.ifftshift(high_pass_result))).astype(np.uint8)
low_pass_image = np.abs(np.fft.ifft2(np.fft.ifftshift(low_pass_result))).astype(np.uint8)

# Salvar as imagens
random_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/random_image.png")
Image.fromarray(high_pass_image).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/high_pass_image.png")
Image.fromarray(low_pass_image).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/low_pass_image.png")

# Mostrar as imagens
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(random_image, cmap='gray')
plt.title('Imagem Aleatória')

plt.subplot(132)
plt.imshow(high_pass_image, cmap='gray')
plt.title('Filtro Passa Alta')

plt.subplot(133)
plt.imshow(low_pass_image, cmap='gray')
plt.title('Filtro Passa Baixa')

plt.tight_layout()
plt.show()
```

```python
# RaY Tracing - 3 esferas

import numpy as np
import matplotlib.pyplot as plt

class Sphere:
    def __init__(self, center, radius, material):
        self.center = center
        self.radius = radius
        self.material = material

    def intersect(self, origin, direction):
        oc = origin - self.center
        a = np.dot(direction, direction)
        b = 2 * np.dot(oc, direction)
        c = np.dot(oc, oc) - self.radius**2
        discriminant = b**2 - 4 * a * c

        if discriminant < 0:
            return None

        t1 = (-b + np.sqrt(discriminant)) / (2 * a)
        t2 = (-b - np.sqrt(discriminant)) / (2 * a)

        if t1 > 0:
            return t1
        elif t2 > 0:
            return t2
        else:
            return None

    def calculate_normal(self, point):
        return (point - self.center) / self.radius

class Material:
    def __init__(self, color):
        self.color = color

    def compute_color(self, point, normal, objects):
        return self.color

class PointLight:
    def __init__(self, position, intensity):
        self.position = position
        self.intensity = intensity


def trace_ray(origin, direction, objects, light_sources):
    closest_t = float('inf')
    closest_object = None

    for obj in objects:
        t = obj.intersect(origin, direction)
        if t is not None and t < closest_t:
            closest_t = t
            closest_object = obj

    if closest_object is None:
        return None

    intersection_point = origin + direction * closest_t
    normal = closest_object.calculate_normal(intersection_point)

    return closest_object.material.compute_color(intersection_point, normal, light_sources)

def render_scene(width, height, camera_pos, objects, light_sources):
    image = np.zeros((height, width, 3))
    aspect_ratio = width / height

    for j in range(height):
        for i in range(width):
            x = (2 * (i + 0.5) / width - 1) * aspect_ratio
            y = 1 - 2 * (j + 0.5) / height

            direction = np.array([x, y, -1])
            direction /= np.linalg.norm(direction)

            color = trace_ray(camera_pos, direction, objects, light_sources)
            image[j, i] = color if color is not None else [0, 0, 0]

    return image

# Definir os materiais
red_material = Material(color=[1, 0, 0])
green_material = Material(color=[0, 1, 0])
blue_material = Material(color=[0, 0, 1])

# Definir as esferas na cena
sphere1 = Sphere(center=[-2, 0, -10], radius=1, material=red_material)
sphere2 = Sphere(center=[0, 0, -10], radius=1, material=green_material)
sphere3 = Sphere(center=[2, 0, -10], radius=1, material=blue_material)

# Definir os objetos na cena
objects = [sphere1, sphere2, sphere3]

# Definir as fontes de luz na cena
light1 = PointLight(position=[-5, 5, 0], intensity=[1, 1, 1])
light2 = PointLight(position=[5, 5, 0], intensity=[1, 1, 1])
light_sources = [light1, light2]

# Definir a posição da câmera
camera_position = np.array([0, 0, 0])

# Renderizar a cena
width = 800
height = 600
image = render_scene(width, height, camera_position, objects, light_sources)

# Exibir a imagem renderizada
plt.imshow(image)
plt.show()
```

```python
# RaY Tracing - Iluminação e sombra

import numpy as np
import matplotlib.pyplot as plt

# Definindo as classes para objetos e luzes
class Sphere:
    def __init__(self, center, radius, color):
        self.center = center
        self.radius = radius
        self.color = color

    def intersect(self, origin, direction):
        oc = origin - self.center
        a = np.dot(direction, direction)
        b = 2 * np.dot(oc, direction)
        c = np.dot(oc, oc) - self.radius**2
        discriminant = b**2 - 4 * a * c

        if discriminant < 0:
            return None

        t1 = (-b + np.sqrt(discriminant)) / (2 * a)
        t2 = (-b - np.sqrt(discriminant)) / (2 * a)

        if t1 > 0:
            return t1
        elif t2 > 0:
            return t2
        else:
            return None

class Light:
    def __init__(self, position, intensity):
        self.position = position
        self.intensity = intensity

# Função para determinar se um ponto está em sombra
def in_shadow(point, objects):
    shadow_ray_direction = light.position - point
    for obj in objects:
        if obj.intersect(point, shadow_ray_direction) is not None:
            return True
    return False

# Função para calcular a cor de um ponto na cena
def trace_ray(origin, direction, objects, lights):
    closest_t = float('inf')
    closest_object = None

    for obj in objects:
        t = obj.intersect(origin, direction)
        if t is not None and t < closest_t:
            closest_t = t
            closest_object = obj

    if closest_object is None:
        return np.array([0, 0, 0])  # Cor de fundo

    intersection_point = origin + direction * closest_t

    # Verifica se o ponto está em sombra
    if in_shadow(intersection_point, objects):
        return np.array([0, 0, 0])  # Ponto está em sombra

    # Calcula a cor difusa usando o modelo de iluminação de Lambert
    normal = (intersection_point - closest_object.center) / closest_object.radius
    light_direction = (light.position - intersection_point) / np.linalg.norm(light.position - intersection_point)
    diffuse_intensity = max(0, np.dot(normal, light_direction))
    diffuse_color = closest_object.color * diffuse_intensity * light.intensity

    return diffuse_color

# Definindo a posição da câmera
camera_pos = np.array([0, 0, 0])

# Definindo a luz
light = Light(position=np.array([2, 2, 2]), intensity=np.array([10, 10, 10]))

# Definindo a esfera na cena (cor vermelha)
sphere = Sphere(center=np.array([0, 0, 0]), radius=1, color=np.array([1, 1, 1]))

# Lista de objetos na cena
objects = [sphere]

# Tamanho da imagem
width = 800
height = 600

# Renderizando a cena
image = np.zeros((height, width, 3))

for j in range(height):
    for i in range(width):
        x = (2 * (i + 0.5) / width - 1)
        y = (1 - 2 * (j + 0.5) / height)
        direction = np.array([x, y, 0]) - camera_pos
        color = trace_ray(camera_pos, direction, objects, [light])
        image[j, i] = np.clip(color, 0, 1)

# Exibindo a imagem renderizada
plt.imshow(image)
plt.show()
```

```python
# transformada e filtragem de imagem aleatória.

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Função para aplicar um filtro passa alta
def high_pass_filter(frequency_image, cutoff_frequency):
    rows, cols = frequency_image.shape
    center_row, center_col = rows // 2 , cols // 2
    mask = np.ones((rows, cols))
    mask[center_row - cutoff_frequency:center_row + cutoff_frequency,
         center_col - cutoff_frequency:center_col + cutoff_frequency] = 0
    return frequency_image * mask

# Função para aplicar um filtro passa baixa
def low_pass_filter(frequency_image, cutoff_frequency):
    rows, cols = frequency_image.shape
    center_row, center_col = rows // 2 , cols // 2
    mask = np.zeros((rows, cols))
    mask[center_row - cutoff_frequency:center_row + cutoff_frequency,
         center_col - cutoff_frequency:center_col + cutoff_frequency] = 1
    return frequency_image * mask

# Gerar uma imagem aleatória de 200x200 pixels
random_image = Image.fromarray(np.random.randint(0, 256, (200, 200), dtype=np.uint8))

# Converter a imagem para um array numpy
image_array = np.array(random_image, dtype=np.float32)

# Aplicar a Transformada de Fourier
frequency_image = np.fft.fftshift(np.fft.fft2(image_array))

# Aplicar filtro passa alta e passa baixa
cutoff_frequency = 30
high_pass_result = high_pass_filter(frequency_image, cutoff_frequency)
low_pass_result = low_pass_filter(frequency_image, cutoff_frequency)

# Aplicar a Transformada Inversa de Fourier para voltar ao domínio do espaço
high_pass_image = np.abs(np.fft.ifft2(np.fft.ifftshift(high_pass_result))).astype(np.uint8)
low_pass_image = np.abs(np.fft.ifft2(np.fft.ifftshift(low_pass_result))).astype(np.uint8)

# Salvar as imagens
random_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/random_image.png")
Image.fromarray(high_pass_image).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/high_pass_image.png")
Image.fromarray(low_pass_image).save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/fourier_transform/low_pass_image.png")

# Mostrar as imagens
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(random_image, cmap='gray')
plt.title('Imagem Aleatória')

plt.subplot(132)
plt.imshow(high_pass_image, cmap='gray')
plt.title('Filtro Passa Alta')

plt.subplot(133)
plt.imshow(low_pass_image, cmap='gray')
plt.title('Filtro Passa Baixa')

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import math

# Define a largura e altura da imagem
width = 800
height = 400

# Cria uma imagem preta
image = Image.new("RGB", (width, height), color=(0, 0, 0))
pixels = image.load()

# Define a posição da câmera
camera_position = (0, 0, -1)

# Define a posição da esfera
sphere_position = (0, 0, 0)
sphere_radius = 0.5

# Define a cor da esfera
sphere_color = (255, 0, 0)

# Define a luz
light_direction = (1, 1, -1)
light_intensity = 1.5

# Loop através de cada pixel na imagem
for x in range(width):
    for y in range(height):
        # Calcula a direção do raio para o pixel atual
        ray_direction = (x / width - 0.5, y / height - 0.5, 1)

        # Normaliza a direção do raio
        ray_direction_length = math.sqrt(ray_direction[0] ** 2 + ray_direction[1] ** 2 + ray_direction[2] ** 2)
        ray_direction = (ray_direction[0] / ray_direction_length, ray_direction[1] / ray_direction_length, ray_direction[2] / ray_direction_length)

        # Calcula a interseção entre o raio e a esfera
        oc = (camera_position[0] - sphere_position[0], camera_position[1] - sphere_position[1], camera_position[2] - sphere_position[2])
        a = ray_direction[0] ** 2 + ray_direction[1] ** 2 + ray_direction[2] ** 2
        b = 2 * (oc[0] * ray_direction[0] + oc[1] * ray_direction[1] + oc[2] * ray_direction[2])
        c = oc[0] ** 2 + oc[1] ** 2 + oc[2] ** 2 - sphere_radius ** 2
        discriminant = b ** 2 - 4 * a * c

        # Se o discriminante for positivo, há uma interseção
        if discriminant > 0:
            # Calcula a cor com base na iluminação
            t = (-b - math.sqrt(discriminant)) / (2 * a)
            intersection_point = (
                camera_position[0] + t * ray_direction[0],
                camera_position[1] + t * ray_direction[1],
                camera_position[2] + t * ray_direction[2]
            )
            normal = (
                intersection_point[0] - sphere_position[0],
                intersection_point[1] - sphere_position[1],
                intersection_point[2] - sphere_position[2]
            )
            normal_length = math.sqrt(normal[0] ** 2 + normal[1] ** 2 + normal[2] ** 2)
            normal = (normal[0] / normal_length, normal[1] / normal_length, normal[2] / normal_length)
            light_dot_normal = light_direction[0] * normal[0] + light_direction[1] * normal[1] + light_direction[2] * normal[2]
            intensity = max(light_dot_normal, 0) * light_intensity
            color = (
                min(int(sphere_color[0] * intensity), 255),
                min(int(sphere_color[1] * intensity), 255),
                min(int(sphere_color[2] * intensity), 255)
            )
            pixels[x, y] = color

# Salva a imagem gerada
image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/ray_tracing_image.png")
```

```python
from PIL import Image, ImageDraw
import random
import math

def add_random_texture(draw, x, y, pattern_size, pattern_type):
    if pattern_type == "balls":
        radius = pattern_size // 4
        draw.ellipse([x - radius, y - radius, x + radius, y + radius], fill=(0, 0, 0))
    elif pattern_type == "hatch":
        hatch_length = pattern_size // 2
        draw.line([x - hatch_length, y, x + hatch_length, y], fill=(0, 0, 0), width=2)
    elif pattern_type == "grid":
        grid_size = pattern_size // 2
        draw.rectangle([x - grid_size, y - grid_size, x + grid_size, y + grid_size], outline=(0, 0, 0))

def add_random_textures(image, pattern_density, pattern_size, pattern_types):
    width, height = image.size
    draw = ImageDraw.Draw(image)

    for x in range(0, width, pattern_size):
        for y in range(0, height, pattern_size):
            if random.random() < pattern_density:
                pattern_type = random.choice(pattern_types)
                add_random_texture(draw, x, y, pattern_size, pattern_type)

# Defina as dimensões da bandeira
width = 600
height = 400

# Crie uma imagem em branco
image = Image.new("RGB", (width, height), color=(255, 255, 255))

# Desenhe as faixas verdes e amarelas
green_height = height // 3
yellow_height = height // 3
blue_height = height // 3

draw = ImageDraw.Draw(image)
draw.rectangle([0, 0, width, green_height], fill=(0, 128, 0))  # Faixa verde
draw.rectangle([0, green_height, width, green_height + yellow_height], fill=(255, 255, 0))  # Faixa amarela
draw.rectangle([0, green_height + yellow_height, width, height], fill=(0, 0, 255))  # Faixa azul

# Adicione texturas aleatórias
pattern_density = 0.2
pattern_size = 40
pattern_types = ["balls", "hatch", "grid"]

add_random_textures(image, pattern_density, pattern_size, pattern_types)

# Salve a imagem gerada
image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/bandeira_brasil_texturas.png")
```

```python
from PIL import Image, ImageDraw

def draw_checkered_flag(width, height, square_size):
    # Crie uma imagem em branco
    image = Image.new("RGB", (width, height), color=(255, 255, 255))
    draw = ImageDraw.Draw(image)

    # Defina as cores das células da bandeira
    colors = [(255, 0, 0), (255, 255, 255)]

    # Desenhe a bandeira quadriculada
    for x in range(0, width, square_size):
        for y in range(0, height, square_size):
            color = colors[(x // square_size + y // square_size) % 2]
            draw.rectangle([x, y, x + square_size, y + square_size], fill=color)

    return image

# Parâmetros da bandeira quadriculada
flag_width = 600
flag_height = 400
square_size = 40

# Desenhe a bandeira quadriculada
flag_image = draw_checkered_flag(flag_width, flag_height, square_size)

# Salve a imagem gerada
flag_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/bandeira_quadriculada.png")
```

```python
from PIL import Image, ImageDraw
import random

def draw_colored_checkered_flag(width, height, square_size):
    # Crie uma imagem em branco
    image = Image.new("RGB", (width, height), color=(255, 255, 255))
    draw = ImageDraw.Draw(image)

    # Defina as cores para as células da bandeira
    colors = [
        (255, 0, 0),  # Vermelho
        (0, 255, 0),  # Verde
        (0, 0, 255),  # Azul
        (255, 255, 0),  # Amarelo
        (255, 165, 0),  # Laranja
        (128, 0, 128)   # Roxo
    ]

    # Desenhe a bandeira quadriculada com cores aleatórias
    for x in range(0, width, square_size):
        for y in range(0, height, square_size):
            color = random.choice(colors)
            draw.rectangle([x, y, x + square_size, y + square_size], fill=color)

    return image

# Parâmetros da bandeira quadriculada colorida
flag_width = 600
flag_height = 400
square_size = 40

# Desenhe a bandeira quadriculada colorida
flag_image = draw_colored_checkered_flag(flag_width, flag_height, square_size)

# Salve a imagem gerada
flag_image.save("/content/drive/MyDrive/Colab Notebooks/arquivos/pdi/bandeira_quadriculada_colorida.png")
```
